# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Figure out the amount of schools. Use unique to alleviate repeats
SchoolName = school_data_complete['school_name'].unique()
#len(SchoolName)

#Find out the total students
TotalStudents = school_data_complete['student_name']
#len(TotalStudents)

#calculate the total budget
#Sum the number once per school using drop_duplicates. inplace: boolean, default False. Whether to drop duplicates in place or to return a copy
TotBudget = school_data_complete.drop_duplicates(subset='budget', keep='first',inplace=False)['budget'].sum()
#TotBudget

#calculate the average math score
AvgMath = school_data_complete['math_score'].mean()
#AvgMath

#calculate the average reading score
AvgRead = school_data_complete['reading_score'].mean()
#AvgRead

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
OverallPass = ((AvgMath + AvgRead) / 2)
#OverallPass

#Calculate the percentage of students with a passing math score (70 or greater)
#Using .loc to access the math score column, only when greater than 70
StudentPass_Math = school_data_complete.loc[school_data_complete['math_score'] >= 70]['math_score'].count()
PP_Math = StudentPass_Math/len(TotalStudents)*100
#PP_Math

#Calculate the percentage of students with a passing reading score (70 or greater)
StudentPass_Read = school_data_complete.loc[school_data_complete['reading_score'] >= 70]['reading_score'].count()
PP_Read = StudentPass_Read/len(TotalStudents)*100
#PP_Read

#Create a dataframe to hold the above results
#Use dictionaries to label column headers 
DistSum_dict = {"Total Schools": [len(SchoolName)],"Total Students": [len(TotalStudents)],"Total Budget": [TotBudget],
                "Average Math Score": [AvgMath],"Average Reading Score": [AvgRead],"% Passing Math": [PP_Math],"% Passing Reading":[PP_Read],
                "Overall Passing Rate": [OverallPass]}

Summary_df = pd.DataFrame(DistSum_dict)


#store as different df to change order
DistSummary = Summary_df[["Total Schools","Total Students","Total Budget","Average Math Score","Average Reading Score",
                      '% Passing Math','% Passing Reading','Overall Passing Rate']]

#Summary
#Optional: give the displayed data cleaner formatting "${:,.2f}".format
DistSummary["Total Students"] = DistSummary["Total Students"].map("{:,}".format)
DistSummary["Total Budget"] = DistSummary["Total Budget"].map("${:,}".format)
DistSummary["Average Math Score"] = DistSummary["Average Math Score"].map("{:.2f}%".format)
DistSummary["Average Reading Score"] = DistSummary["Average Reading Score"].map("{:.2f}%".format)
DistSummary["% Passing Math"] = DistSummary["% Passing Math"].map("{:.2f}%".format)
DistSummary["% Passing Reading"] = DistSummary["% Passing Reading"].map("{:.2f}%".format)
DistSummary["Overall Passing Rate"] = DistSummary["Overall Passing Rate"].map("{:.2f}%".format)

DistSummary
```



## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate
* use the following headers for a table:
* School Type, Total Students, Total School Budget, Per Student Budget, Average Math Score, Average Reading Score, % Passing Math, % Passing Reading, % Overall Passing Rate

* merged table columns are:
* Student ID	student_name	gender	grade	school_name	reading_score	math_score	School ID	type	size	budget

```python
#group by the school name
#Group series using mapper (dict or key function return result as series) or by a series of columns.
School_Grouped_df = student_data.groupby(["school_name"], as_index=False)
#School_Grouped_df.head()

# Math scores (average) per school
AvgMath_df = pd.DataFrame(School_Grouped_df["math_score"].mean())
#AvgMath_df.head()

#Find math scores better than 70%
MathPass = student_data[student_data["math_score"]>70].groupby(["school_name"], as_index=False)
#MathPass.head()

# total math pass scores as a prep for the percent pass
MathPass_df = pd.DataFrame(MathPass["math_score"].count())
MathPass_df.columns= ["school_name", "% Math Pass"]
#MathPass_df.head()

#  Readings scores (average) per school
AvgReading_df = pd.DataFrame(School_Grouped_df["reading_score"].mean())
#AvgReading_df.head()

#Find reading scores better than 70%
ReadingPass = student_data[student_data["reading_score"]>70].groupby(["school_name"], as_index=False)
#ReadingPass.head()

# total reading pass scores as a prep for the percent pass
ReadPass_df = pd.DataFrame(ReadingPass["reading_score"].count())
ReadPass_df.columns= ["school_name", "% Reading Pass"]
#ReadPass_df

# Merge school file with grouped data
School_Merged_df = pd.merge(school_data, AvgMath_df, on="school_name" )
School_Merged_df = pd.merge(School_Merged_df, MathPass_df, on="school_name" )
School_Merged_df = pd.merge(School_Merged_df, AvgReading_df, on="school_name" )
School_Merged_df = pd.merge(School_Merged_df, ReadPass_df, on="school_name" )
#School_Merged_df.head()

# Calculate math and reading percent pass within each school
School_Merged_df["% Math Pass"] = (School_Merged_df["% Math Pass"]/School_Merged_df["size"])*100
School_Merged_df["% Reading Pass"] = (School_Merged_df["% Reading Pass"]/School_Merged_df["size"])*100
#School_Merged_df

# Calculate Overall Percent Passing
School_Merged_df["Overall Percent Pass"] = (School_Merged_df["% Math Pass"]+School_Merged_df["% Reading Pass"])/2
School_Merged_df

# Find the per student Budget and add column
School_Merged_df["Budget per Student"] = School_Merged_df["budget"]/School_Merged_df["size"]
School_Merged_df

# Select and rename columns
Top5_School_summary = pd.DataFrame(School_Merged_df[["school_name", "type", "size", "budget",                                                     "Budget per Student", "math_score", "reading_score", 
                                                     "% Math Pass", "% Reading Pass", "Overall Percent Pass"]])
#Top5_School_summary
Top5_School_summary.columns = ["School Name", "Type", "Num Students", "Total Budget", "Budget per Student", 
                               "Avg Math Score", "Avg Reading Score", "% Math Pass", "% Reading Pass", "Overall Percent Pass"]

# Apply formatting to columns
Top5_School_summary["Total Budget"] = Top5_School_summary["Total Budget"].map("${:,}".format)
Top5_School_summary["Budget per Student"] = Top5_School_summary["Budget per Student"].map("${:,}".format)
Top5_School_summary["Avg Math Score"] = Top5_School_summary["Avg Math Score"].map("{:.2f}".format)
Top5_School_summary["Avg Reading Score"] = Top5_School_summary["Avg Reading Score"].map("{:.2f}".format)
Top5_School_summary["% Math Pass"] = Top5_School_summary["% Math Pass"].map("{:.2f}%".format)
Top5_School_summary["% Reading Pass"] = Top5_School_summary["% Reading Pass"].map("{:.2f}%".format)
Top5_School_summary["Overall Percent Pass"] = Top5_School_summary["Overall Percent Pass"].map("{:.2f}%".format)
#Top5_School_summary

# Sort and list top 5. Ascending = False sorts descending.  Take the first 5 of the list
Top5_School_sort = Top5_School_summary.sort_values("Overall Percent Pass", ascending=False)[:5]
Top5_School_sort.head(10)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#create grade level average math scores for each school 
NinethG_math = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
TenthG_math = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
EleventhG_math = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()
TwelthG_math = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()

# add math scores to a dictionary
math_scores_per_school = pd.DataFrame({"9th": NinethG_math,"10th": TenthG_math,"11th": EleventhG_math,"12th": TwelthG_math})
math_scores_per_school = math_scores_per_school[['9th', '10th', '11th', '12th']]
math_scores_per_school.index.name = "School"

#show and format
math_scores_per_school.style.format({'9th': '{:.1f}',"10th": '{:.1f}',"11th": "{:.1f}","12th": "{:.1f}"})
```
